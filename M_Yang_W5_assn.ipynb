# 1.which dayofweek has the most flights to fly in January?
 The operater of airport want to know which day of week is the busiest that has the most of flights fly.For this question, I extract the January flights' information first. I counts the times of flights flying in January for each dayofweek.Then,I create the summary statistic tabke and histogram to analysis the question. Acoording to the result, they can know which day of week should arrange most employees to offer customer services. 
 ###extract the January flights from hflights
 ###create the new datafarm that contains the dayofweek's group and times of flights flying in January.
 ###create the summary statistic datafarm that relates to the dayofweek in grouping and counting.
 ###create the pandas graphic,frequency bar chart,to check the if there more than one dayofweek has the maximum times of flights flying in January.
 ###create the datafarm show the dayofweek that has the maximum times of flights flying in January.

```python
#find all January flights, which is the Month = 1
#use groupby each dayofweek and count the number for each dayofweek in January
January_flights = hflights[(hflights['Month'] == 1)]
weekday_January_flights = January_flights.groupby('DayOfWeek')[['DayOfWeek']].count()
```

```python
#Utilize the describe function to get the basic statistic of the day of week in Junuary, we can find the max values of flights' flying time.
weekday_January_flights.describe()
```

```python
#find the frequency of max value of flights flying times in January.According to the frequency by histigram, we can know there are no two different dayofweek have same max value. 

weekday_January_flights['DayOfWeek'].plot(kind = 'hist')
```

```python
#find the Dayofweek whose count number is 3333 in weekday_January_flights table. 
weekday_January_flights[weekday_January_flights['DayOfWeek'].isin([3333])]
# The Monday is the day that airport have the most flights fly in January. (Assume 1 means Monday)
```
