```python
import pandas as pd
import numpy as np
# read the data file from GitHub and convert to a NumPy ndarray object
filename = "https://raw.githubusercontent.com/jtopor/DAV-5400/master/Project1/hflights.csv" 
hflights = pd.read_csv(filename)
hflights
hflights.describe()
```

#1.which dayofweek has the most flights to fly in January?

```python
#find all January flights
#count the number for each dayofweek in January
January_flights = hflights[(hflights['Month'] == 1)]
weekday_January_flights = January_flights.groupby('DayOfWeek')[['DayOfWeek']].count()
max1 = weekday_January_flights['DayOfWeek'].max()
print(max1)
```
